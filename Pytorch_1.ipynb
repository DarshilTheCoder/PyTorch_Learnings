# Tensor Maths Operation

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])

print(tensor_a,tensor_b)
```

```python
#Addition
print(tensor_a+tensor_b)
#or
#torch.add(tensor_a,tensor_b)
#or
tensor_a.add(tensor_b)
```

```python
#Subtraction
print(tensor_a-tensor_b)
#OR
torch.sub(tensor_a,tensor_b)
```

```python
#Multiplication
print(tensor_a*tensor_b)
#or
torch.mul(tensor_a,tensor_b)
```

```python
#Division
print((tensor_a/tensor_b))
#or
torch.div(tensor_a,tensor_b)
```

```python
#Remainders
print(tensor_a % tensor_b)
print(tensor_b % tensor_a)
#or
torch.remainder(tensor_b,tensor_a)
```

```python
#Reassingment
tensor_c = tensor_a + tensor_b
print(tensor_c)
#or
print(tensor_a-tensor_b)
tensor_a.add_(tensor_a-tensor_b) #what here is happening is that 1+(1-5=-4) so 1-4=-3 is assigning to tensor_a and that's why getting -3, similarly for second one 2-6=-4 and 2-4 = -2.
tensor_a
```
